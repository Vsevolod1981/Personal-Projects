## Сведем данные за 4 месяца в один датафрейм.

```python
# Преобразуем данные из столбца 'Дата' в формат datatime для каждого месяца.

march['Дата'] = march['Дата'].apply(lambda x: x.split(','))
march['Время'] = march['Дата'].apply(lambda x: x[1])
march['Дата'] = march['Дата'].apply(lambda x: x[0])
march['Дата'] = march['Дата'].apply(lambda x: x.split(' '))
march['Дата'] = march['Дата'].apply(lambda x: x[0])
march['Дата'] = march['Дата'].apply(lambda x: x+'.03.2022')
march['Дата'] = march['Дата'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y'))

april['Дата'] = april['Дата'].apply(lambda x: x.split(','))
april['Время'] = april['Дата'].apply(lambda x: x[1])
april['Дата'] = april['Дата'].apply(lambda x: x[0])
april['Дата'] = april['Дата'].apply(lambda x: x.split(' '))
april['Дата'] = april['Дата'].apply(lambda x: x[0])
april['Дата'] = april['Дата'].apply(lambda x: x+'.04.2022')
april['Дата'] = april['Дата'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y'))

may['Дата'] = may['Дата'].apply(lambda x: x.split(','))
may['Время'] = may['Дата'].apply(lambda x: x[1])
may['Дата'] = may['Дата'].apply(lambda x: x[0])
may['Дата'] = may['Дата'].apply(lambda x: x.split(' '))
may['Дата'] = may['Дата'].apply(lambda x: x[0])
may['Дата'] = may['Дата'].apply(lambda x: x+'.05.2022')
may['Дата'] = may['Дата'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y'))

july['Дата'] = july['Дата'].apply(lambda x: x.split(','))
july['Время'] = july['Дата'].apply(lambda x: x[1])
july['Дата'] = july['Дата'].apply(lambda x: x[0])
july['Дата'] = july['Дата'].apply(lambda x: x.split(' '))
july['Дата'] = july['Дата'].apply(lambda x: x[0])
july['Дата'] = july['Дата'].apply(lambda x: x+'.07.2022')
july['Дата'] = july['Дата'].apply(lambda x: datetime.strptime(x, '%d.%m.%Y'))
```

```python
# Название столбцов во всех датасетах одинаковые, поэтому подклеим один за одним снизу, предварительно в каждом создав столбец
# 'Месяц'.
march['Месяц'] = 'Март'
april['Месяц'] = 'Апрель'
may['Месяц'] = 'Май'
july['Месяц'] = 'Июль'

all_period = march.append(april,  ignore_index=True)
all_period = all_period.append(may,  ignore_index=True)
all_period = all_period.append(july,  ignore_index=True)

all_period
```
